## ARITHMETIC

```python
num1=int(input("enter first number"))
num2=int(input("enter second number"))
print("the sum is",num1+num2)
print("the sub is",num1-num2)
print("the multi is",num1*num2)
print("the exp is",num1**num2)             #10 power 25
print("the div is",num1/num2)
print("the floor div is",num1//num2)       # is similar to division but removes the decimal part by giving the small nearest number

print("the reminder is",num1%num2)
```

## RELATIONAL

```python
a=int(input("value of a"))
b=int(input("value of b"))
print("the equal to is",a==b)
print("the not equal to is",a!=b)
print("the greater than is",a>b)
print("the smaller than is",a<b)
print("the smaller than equal to is",a<=b)
```

## LOGICAL

- and  takes atleast 2 input and even if one is false then answer is also false
